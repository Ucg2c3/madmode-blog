# Performance Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Data...

```python
m = 60
index = pd.Series(range(1, 11), range(1, 11))
rktimes = pd.Series([103.13, 109.59, 188.52, 236.11, 348.45, 545.95,
                     13 * m + 15.53, 20 * m + 14.12, 31 * m + 51.09, 49 * m + 57.72],
                    index=index)

bstimes = pd.Series([70, 81.51, 137.07, 134.87, 193.85, 198.69, 218.22, 298.71, 298.42, 297.06],
                    index=index)

tolerance = pd.Series(index, index)
```

## Table of tolerance versus time

```python
both_times = pd.DataFrame(dict(RKQC=rktimes, BSSTEP=bstimes))
both_times
```

## Graphs of tolerance versus time

```python
%matplotlib inline
```

```python
def bench_plot(s, title):
    fig = s.plot()
    fig.set_title(title)
    fig.set_xlabel("log(1/eps)")
    fig.set_ylabel("time(sec)")
    return fig
```

```python
bp = bench_plot(bstimes, "Bulirsch-Stoer Times")
```

```python
rp = bench_plot(rktimes, "Runge-Kutta Times")
```

```python
_ = bench_plot(both_times, "Both Times")
```

## Graph of tolerance versus relative speed

```python
plt.title("Relative Speed: BS:RK")
plt.xlabel("log(1/eps)")
plt.ylabel("Percent")
_ = (rktimes / bstimes * 100).plot()
```

## Fit of time data to curves

```python
from scipy.optimize import curve_fit

import sympy
from sympy.abc import x
from sympy.plotting import plot as symplot

sympy.init_printing()
```

```python
def fit(data, f):
    popt, pcov = curve_fit(f, index, data)
    poly = sum(round(c, 4) * x ** i for (i, c) in enumerate(popt))
    return poly

at = lambda f, s: sympy.lambdify(s, f)
```

```python
bfit = fit(bstimes, lambda x, c0, c1: c0 + c1 * x)
rfit = fit(rktimes, lambda x, c0, c1, c2, c3: c0 + c1 * x + c2 * x**2 + c3 * x**3)
_ = at(bfit, x)(index).plot()
_ = at(rfit, x)(index).plot()
```
